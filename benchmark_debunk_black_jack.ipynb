- Trước hết, tôi sẽ cho các bạn xem biểu đồ thắng / thua / hoà của nhà cái đối với người chơi. Do số lượng mẫu quá nhiều, tôi sẽ sử dụng Monto Carlo để giả lập 5 triệu lượt chơi.

```python
# Thực hiện import các thư viện cần thiết
import debunk_blackjack
db = debunk_blackjack.BlackJack()
```

- Bây giờ, ta cho chạy để lấy biểu đô thắng, thua, hoà của xì dách. Trong lần này, ta sẽ cho chạy dưới điều kiện là 1 người chơi đấu với 1 nhà cái, và cả hai sẽ rút đến khi đạt 17 điểm hoặc hơn thì ngừng.

```python
db.Monto_carlo_black_jack(turns=5000000, vizualize=True)
```

- Theo biểu đồ, ta thấy rằng từ đoạn 4 đến 6, tỉ lệ chiến thắng của nhà cái giảm dần từ dưới 50% đến 40%, từ đoạn 7 đến 12, tỉ lệ chiến thắng của nhà cái tăng từ dưới 40% lên hơn 60%. Từ đoạn 12 - 16, tỉ lệ chiến thắng của nhà cái cực kì thấp, luôn nằm dưới mức 40% và chạm đến 30%. Cuối cùng, ta thấy từ đoạn 17 đi lên, tỉ lệ tăng đột biến, cao nhất là 20 điểm với hơn 60% tỉ lệ thắng.
- Ngoài ra, tỉ lệ chiến thắng của nhà cái cực kì cao lên đến 43.87%, trong khi đó tỉ lệ hoà là 17.60%, tỉ lệ thua là 38.52%, điều này chứng minh cho việc nhà cái luôn có tỉ lệ thắng cao hơn so với nhà cái.
- Tỉ lệ net hiện tại là -0.00995%, tức là mỗi cuộc chơi thì nhà cái sẽ mất 0.0001 giá trị, gần như xấp xỉ là 0. Tuy vậy, nếu như tăng số lượng người chơi, thì kết quả sẽ rất khác.
- Ta thực hiện giả lập 5 người chơi đấu 5 triệu trận với nhà cái và cho kết quả.

```python
db.Monto_carlo_black_jack(player = 5, turns = 5000000, vizualize=True)
```

Ta thấy rằng, tỉ lệ thắng/thua của nhà cái đã tăng lên đáng kể, từ 43.85% -> 49.74%. Nhưng quan trong hơn, tỉ lệ net hiện tại là 0.15%, tức là sau một ván nhà cái sẽ nhận được 0.0015 giá trị. Chỉ số này đã tăng gấp 15 lần so với khi chơi 1 người. Điều này chứng tỏ rằng, đây là trò chơi không công bằng và phần lợi thế nghiêng về nhà cái.

Tiếp theo, ta sẽ kiểm tra xem, trong một trận đấu 1vs1 với nhà cái, ta sẽ có bao nhiêu phần trăm đạt được mục tiêu mong muốn.
Ta sẽ cho số vốn ban đầu là 100k, số tiền mỗi lần cược là 10k và mục tiêu lần lượt là 150k, 200k, 250k, 300k. Với mỗi mục tiêu, ta sẽ thực hiện 10000 lần để kiểm tra tỉ lệ thắng của người chơi.

```python
target = [150000, 200000, 250000, 300000]
trials = 10000
for t in target:
    wins = 0
    for _ in range(trials):
        if db.combat1vs1(target=t):
            wins += 1
    print(f"target={t}, win_fraction={wins}/{trials} => {wins/trials:.4f}")
```

Như ta có thể thấy ở kết quả trên, tỉ lệ chiến thắng của người chơi thay đổi theo mục tiêu của họ đề ra. Với mục tiêu là 150000, tỉ lệ có thể đạt được là 63.82%, nhưng khi mục tiêu đạt mức gấp đôi, tỉ lệ đạt được giảm xuống dưới 50% cho thấy rằng càng đề ra mục tiêu càng cao thì khả năng thất bại càng lớn.
